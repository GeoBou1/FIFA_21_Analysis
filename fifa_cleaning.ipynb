### Cleaning

Let's start by correcting and renaming our incorrect columns.

```python
fifa_data_fitered
```

```python
#Renaming few columns

fifa_data_fitered=fifa_data_fitered.rename(columns={
    '↓OVA':'Overall'
    ,'Best Position':'Best_Position'
    ,'Release Clause':'Release_Clause'
})

fifa_data_fitered
```

```python
#Getting rid off the typo 

fifa_data_fitered['Club']=fifa_data_fitered['Club'].str.strip('\n\n\n\n')

fifa_data_fitered
```

```python
#Getting rid off the measurement units and convert from str to int

fifa_data_fitered["Height"]=fifa_data_fitered["Height"].str.strip("cm")
fifa_data_fitered["Height"].unique()

def ft_to_cm(x):
    if "'" in x:
        ft=x.replace("'",".").replace('"',"")
        cm= round(float(ft)*30.48,0)
        return int(cm)

    else:
        return int(x)

fifa_data_fitered["Height"]=fifa_data_fitered["Height"].apply(ft_to_cm)
fifa_data_fitered["Height"].unique()
```

```python
#Getting rid off the measurement units and convert from str to int

fifa_data_fitered["Weight"]=fifa_data_fitered["Weight"].str.strip("kg")
fifa_data_fitered["Weight"].unique()

def lbs_to_kg(x):
    if "lbs" in x:
        lbs = x.replace("lbs", "")
        kilograms = round(int(lbs) / 2.2, 0)
        return int(kilograms)
    else:
        return int(x)


fifa_data_fitered["Weight"]=fifa_data_fitered["Weight"].apply(lbs_to_kg)
fifa_data_fitered["Weight"].unique()
```

```python
#Renaming columns with units

fifa_data_fitered=fifa_data_fitered.rename(columns={
    'Height':'Height_Cm'
    ,'Weight':'Weight_Kg'
})

fifa_data_fitered
```

```python
def money(x):

    x=str(x)

    if "€" in x:
        x = x.replace("€", "")

    if "M" in x:
        x = x.replace("M", "")
        return int(float(x) * 1_000_000)

    elif "K" in x:
        x = x.replace("K", "")
        return int(float(x) * 1_000)

    return int(x)

fifa_data_fitered["Value"] = fifa_data_fitered["Value"].apply(money)
fifa_data_fitered["Wage"] = fifa_data_fitered["Wage"].apply(money)
fifa_data_fitered["Release_Clause"] = fifa_data_fitered["Release_Clause"].apply(money)

fifa_data_fitered
```

The dataset looks a bit cleaner now ! The last step will us the opportunity to look at the fidelity of players : we'll make a new column, called 'In_Club' in which we'll be able to see for how many years a player has been in a club. To simplify, we'll just look at years.

```python
fifa_data_fitered['Joined']=fifa_data_fitered['Joined'].str.strip().str[-4:]
```

```python
fifa_data_fitered['Joined'].value_counts()
```

```python
fifa_data_fitered['Joined']=fifa_data_fitered['Joined'].astype(int)

fifa_data_fitered.info()
```

Assuming we're in 2021, we'll create a new column.

```python
fifa_data_fitered['In_Club']=2021-fifa_data_fitered['Joined']

fifa_data_fitered
```

## Data Analysis

### Distribution on different parameters

```python
import plotly.express as px
import seaborn as sns
```

```python
fifa_data_fitered.groupby(['Nationality'])['ID'].count()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

nationality_counts = fifa_data_fitered.groupby('Nationality')['ID'].count().reset_index()

nationality_counts = nationality_counts.sort_values(by='ID', ascending=False)

nationality_counts = nationality_counts.head(10)

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Age Distribution", "Nationalities Top 10"),
    specs=[[{'type': 'xy'}, {'type': 'domain'}]]
)

fig.add_trace(
    go.Histogram(
        x=fifa_data_fitered['Age'],
        y=fifa_data_fitered['ID'],
        histfunc='count',
        showlegend=True
    ),
    row=1, col=1
)

fig.add_trace(
    go.Pie(
        labels=nationality_counts['Nationality'],
        values=nationality_counts['ID'],
        name="Nationalité",
        pull=[0.1] * len(nationality_counts),
        showlegend=True
    ),
    row=1, col=2
)


fig.update_layout(
    title_text="Age Distirbution and Nationalities",
    showlegend=True,
)

fig.show()
```

```python
print('The average height of players is '+str(round(fifa_data_fitered['Height_Cm'].mean(),2))+' cm.')
print('The average weight of players is '+str(round(fifa_data_fitered['Weight_Kg'].mean(),2))+' kg.')
print('The average age of players is '+str(round(fifa_data_fitered['Age'].mean(),2))+' years old.')
```
